```python
# stampare dict con chiave il bigramma e valore una lista dei caratteri successivi a quel bigramma
with open("OSCAR_WILDE_Il_fantasma_di_Canterville.txt", "+r") as file_to_read:
    lines_of_file = file_to_read.read()
last_char = ''
number_of_bigrams = {}
for char in lines_of_file:
    bigram = last_char+char
    last_char = char
    if bigram in number_of_bigrams:
        number_of_bigrams[bigram] += 1
    else:
        number_of_bigrams[bigram] = 1
print(number_of_bigrams)
```

```python
n_char_ngram = 7
ngram_dict = {}

for i in range(len(lines_of_file) - n_char_ngram):
    ngram = lines_of_file[i:i+n_char_ngram] 
    next_char = lines_of_file[i+n_char_ngram] 
    
    if ngram in ngram_dict:
        if next_char not in ngram_dict[ngram]:
            ngram_dict[ngram].append(next_char)
    else:
        ngram_dict[ngram] = [next_char]

ngram_dict
        
```

# AUTOREGRESSIONE O PROCESSO DI MARKOV

```python
# chiedere un bigramma e sceglierne uno a caso

import random
N_CHAR = 300
user_input = ""
phrase = ""
while user_input not in ngram_dict:
    user_input = input("Scegli il bigramma da continuare: ")
phrase += user_input

for i in range(0, N_CHAR):
    ngram = phrase[-n_char_ngram:]
    elem = random.choice(ngram_dict[ngram] )
    phrase = phrase+elem
phrase
```
